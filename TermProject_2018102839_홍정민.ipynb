# 특수교육 대상자들이 사는 지역에 따라 교육 격차가 존재하는가?
파이선을 활용한 데이터 저널리즘 프로젝트

2018102839 홍정민

## 1. 프로젝트 목표 및 내용

이 프로젝트의 목표는 특수교육 대상자들이 질 좋은 교육을 받기 위해 감수해야 하는 불편함에 주목해보고, 이를 개선하기 위한 방법을 제시하는 것입니다. 파이선을 이용해 특수교육 대상자들이 겪는 지역별 교육 격차가 존재하는지, 존재한다면 격차가 얼마나 벌어져 있는지 살펴봄으로써 데이터를 기반으로 한 주장과 근거를 제시하려고 합니다.

그렇다면 특수 교육의 질에 영향을 주는 것은 무엇이 있을까요? 저는 개개인의 교사의 노력도 중요하지만, 주변의 환경적인 요소도 교육의 질에 적지 않은 영향을 줄 수 있다고 생각합니다. 교사의 업무 과중 현상, 대중교통 이용의 어려움 등이 학생과 교사들을 심리적으로 지치게 할 수 있고, 학교의 재정적 지원이 부족하다면 특수 교육에 필요한 여러 기구를 구매할 수 없기 때문입니다. 따라서 다음의 요인들을 정하고 이를 지역별로 비교해봄으로써 지역별 차이가 존재하는지, 만약 존재한다면 그 이유는 무엇인지 파악해보려고 합니다.

- 특수 교사 1인당 학생 수
- 장애인 편의 시설의 설치 여부
- 지역별 특수 교육 대상자의 수 대비 특수교육을 제공하는 기관의 수
- 상담 교사 배치 여부

## 2. 주제 선정 이유

대한민국은 현재 교육 기본법 제 18조와 장애인 등에 대한 특수 교육법을 통해 특수 교육 대상자들이 각자의 특성에 맞는 교육을 받을 수 있도록 교육권을 법적으로 보장하고 있습니다. 하지만 특수 교육을 필요로 하는 학생들은 전문적인 교사의 지원과 특수 시설이 필요함에도 불구하고 경희대학교 외국어대학관과 같이 교내에 엘리베이터나 경사로가 설치되지 않는 등 장애 학생들의 학습을 지원할 수 있는 기본적인 시설도 갖춰져 있지 않는 경우가 많아 비장애인 학생들에 비해 학업에 큰 어려움이 있습니다. 
    

**제18조** <br>
<code>(특수교육) 국가와 지방자치단체는 신체적ㆍ정신적ㆍ지적 장애 등으로 특별한 교육적 배려가 필요한 자를 위한 학교를 설립ㆍ경영하여야 하며, 이들의 교육을 지원하기 위하여 필요한 시책을 수립ㆍ실시하여야 한다. </code>

저는 이번 학기에 특수교육학개론 수업을 들으면서 특수교육대상자들에 대해 관심을 갖게 되었고, 그들에게 좀 더 개별화된 교육이 필요하는 것을 배웠습니다. 비장애인 학생들을 대상으로 하는 교육과 달리 필요한 시설과 도구도 많고, 특수교사와의 긴밀한 협업과 전문적인 지원을 위한 선별 검사 과정이 필요하다는 것도 알게 되었습니다. 하지만 최근 들어서 코로나 19로 인한 비대면 수업에서 자막이 제공되지 않아 학업의 어려움을 겪는 대학생에 대한 뉴스를 접했고, 장애를 가진 몇몇 학생들은 아직도 교육적 지원에 있어 배제되어 있을 수 있다고 생각했습니다. 비장애인 학생들도 지역에 따라서 가정환경에 따라서 교육의 질이 천차만별이므로, 어쩌면 장애 아동들은 거주 지역에 따라서 더 큰 불편을 겪을 수 있을 거라 생각했습니다.

또한, 학교 알리미에서 제공하는 '학년별,학급별 학생수'와 '직위별 교원 현황 통계'를 이용해 특수 교사 1인 당 학생 수를 계산하려 합니다.

## 4. 구현 계획

Pandas를 이용하여 지역과 특수 교육 지원 정도에 있어 상관 관계가 있는지 파악하고, folium 모듈을 이용해 지도를 행정구역별로 나눈 후 지도에 특수 교육 지원 정도를 시각화하여 나타내고 싶습니다. 앞서 언급한 모듈들과 Pandas에 대한 사전 지식이 없기 떄문에 중간 보고서 이전까지 이에 대해 공부해보록 하겠습니다.
